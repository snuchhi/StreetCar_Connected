```python
def travel_time(df):
    travel_time = df.groupby('temporaryId')
    traveltime = (pd.concat([travel_time.head(1), travel_time.tail(1)])
               .sort_values('temporaryId').sort_index()
               .reset_index(drop=True))
    traveltime['timestamp_verbose'] = pd.to_datetime(traveltime['timestamp_verbose'])
    traveltime['traveltime'] = traveltime['timestamp_verbose'].iloc[1::2] - traveltime['timestamp_verbose'].iloc[::2].to_numpy()
    df_traveltime = traveltime[np.arange(len(traveltime)) % 2 != 0]
    df_traveltime = df_traveltime[['temporaryId', 'traveltime','distance_along_path']]
    return df_traveltime.style.hide_index()
```

```python
def haversine_in(lat1, lon1,lat2,lon2, to_radians=True, earth_radius=6371):
    """
    slightly modified version: of http://stackoverflow.com/a/29546836/2901002

    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees or in radians)

    All (lat, lon) coordinates must have numeric dtypes and be of equal length.

    """
    
    if to_radians:
        lat1, lon1, lat2, lon2 = np.radians([lat1, lon1, lat2, lon2])

    a = np.sin((lat2-lat1)/2.0)**2 + \
        np.cos(lat1) * np.cos(lat2) * np.sin((lon2-lon1)/2.0)**2

    return earth_radius * 2 * np.arcsin(np.sqrt(a)) * 1000
```
